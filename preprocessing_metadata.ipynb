```python
data = pd.read_csv('data/fashion.csv')
data
```

```python
print(data["SubCategory"].unique())
print(data["ProductType"].unique())
print(data["Usage"].unique())
```

```python
import pandas as pd
import os

IMAGE_DIR = "data/images"  

df = data
def norm_text(x):
    if pd.isna(x):
        return x
    return str(x).strip().lower()

for col in [
    "Gender",
    "Category",
    "SubCategory",
    "ProductType",
    "Colour",
    "Usage"
]:
    df[col] = df[col].apply(norm_text)

df["ProductTitle"] = df["ProductTitle"].apply(lambda x: str(x).strip())
df["image_path"] = df["Image"].apply(lambda x: os.path.join(IMAGE_DIR, str(x)))

df = df.rename(columns={
    "ProductId": "product_id",
    "Gender": "gender",
    "Category": "category",
    "SubCategory": "sub_category",
    "ProductType": "product_type",
    "Colour": "colour",
    "Usage": "usage",
    "ProductTitle": "title",
    "ImageURL": "image_url"
})
df.drop(columns=["Image"], inplace=True)
df["product_type"] = df["product_type"].apply(lambda x: x.split()[-1])
df.to_csv("metadata_preprocessed.csv", index=False)
df
```
